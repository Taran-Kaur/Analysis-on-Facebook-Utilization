```python
data[data['tenure'].isnull() == True]
```

**Observation:**
- We can observe there are missing values in 2 columns - Gender & Tenure
- Gender has 175 missing values
- Tenure has 2 missing values

|Feature|Object Type||Solution|
|:--:|:--:|:--:|:--|
|Gender|object||Replace with Trans
|Tenure|float64|| Replace with 25% value, as we dont know how much the user is active on fb so lets assume 20%-25% of the time to be active

**Performing Operations**

```python
data['tenure'].median()/2
```

```python
#SOLVING FOR MISSING VALUES
#replacing missing value in Tenure column with its 25% value

data['tenure'] = data['tenure'].replace(np.nan,data['tenure'].median()/2)
```

```python
data[data['tenure'].isnull() == True]
```

```python
# So now tenure doesn't have any missing values.
data.shape[0]-data.count()
```

```python
# before change to Nan value
data['gender'].unique()
```

```python
# Replace NaN values for a gender column
data['gender'].fillna('Trans',inplace = True)
```

```python
data['gender'].unique()
```

```python
# So now gender doesn't have any missing values.
data.shape[0]-data.count()
```
